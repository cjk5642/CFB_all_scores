```python
#import packages
import pandas as pd
import html5lib
import numpy as np
```

```python
#define list of concatentation
all_seasons = []

#define loop
for i in range(84):
   #define base year
    year = 1935 + i
    
    #if statement for ties
    if year < 1996:
        
        #find link to data
        url = f"https://www.sports-reference.com/cfb/years/{year}-ratings.html"
        season = pd.read_html(url)

        #get the table
        season = pd.DataFrame(season[0])
        
        #remove last two columns to get rid of totals
        season.drop(season.columns[[-1, -2]], axis = 1, inplace = True)
        
        
        #impose names
        season.columns = ['Rk', 'School', 'Conference', 'AP Rank', 'Wins', 'Loses',
                         'Ties', 'OSRS', 'DSRS', 'SRS', 'Offensive Scoring', 
                          'Defensive Scoring', 'Offensive Passing', 'Defensive Scoring',
                         'Offensive Rushing', 'Defensive Rushing']

        #remove the unneccesary rows
        season = season[season.Conference != 'Overall'][season.Rk != 'Rk']
        
        #add year column
        season['Year'] = str(year)
        
        #reorder columns
        season = season[['Rk', 'School', 'Conference', 'AP Rank', 'Year', 'Wins', 'Loses', 'Ties',
                          'OSRS', 'DSRS', 'SRS', 'Offensive Scoring','Defensive Scoring',
                          'Offensive Passing', 'Defensive Scoring','Offensive Rushing',
                          'Defensive Rushing']]
        
        #remove duplicate columns
        season = season.loc[:,~season.columns.duplicated()]

        #create new dataframe and append all frames
        all_seasons.append(season)
               
    else:
        #find link to data
        url = "https://www.sports-reference.com/cfb/years/" + str(year) + "-ratings.html"
        season = pd.read_html(url)

        #get the table
        season = pd.DataFrame(season[0])

        #remove last two columns to get rid of totals
        season.drop(season.columns[[-1, -2]], axis = 1, inplace = True)
    
        
        #impose names
        season.columns = ['Rk', 'School', 'Conference', 'AP Rank', 'Wins', 'Loses',
                          'OSRS', 'DSRS', 'SRS', 'Offensive Scoring','Defensive Scoring',
                          'Offensive Passing', 'Defensive Scoring','Offensive Rushing',
                          'Defensive Rushing']
        
        #add ties column
        season['Ties'] = '0'

        #remove the unneccesary rows
        season = season[season.Conference != 'Overall'][season.Rk != 'Rk']
        
        #add year column
        season['Year'] = str(year)
        
        #reorder columns
        season = season[['Rk', 'School', 'Conference', 'AP Rank', 'Year', 'Wins', 'Loses', 'Ties',
                          'OSRS', 'DSRS', 'SRS', 'Offensive Scoring','Defensive Scoring',
                          'Offensive Passing', 'Defensive Scoring','Offensive Rushing',
                          'Defensive Rushing']]
        
        #remove duplicate columns
        season = season.loc[:,~season.columns.duplicated()]

        #create new dataframe and append all frames
        all_seasons.append(season)
```

```python
#concatenate all dataframes into on
df = pd.concat(all_seasons, axis=0, ignore_index=True, sort = False)
```

```python
df = df.fillna(0)

conference = sorted(list(set(df.Conference)))
school = sorted(list(set(df.School)))

conf_factor = {c:f for f, c in enumerate(conference)}
school_factor = {s: f for f, s in enumerate(school)}

df['Conference'] = [conf_factor[df.loc[row, 'Conference']] for row in range(len(df))]
df['School'] = [school_factor[df.loc[row, 'School']] for row in range(len(df))]

df = df.astype(np.float64)
display(df)
#convert to csv
#df.to_csv(r'C:\Users\colli\FOLDER OF ALL\Desktop\Projects\Data for College Football\CFBData.csv', encoding='utf-8')
print('File sent successfully.')
```
